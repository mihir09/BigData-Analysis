```python
import os
import sys
from dateutil.relativedelta import relativedelta
from datetime import datetime
```

```python
%%time
res_limit = 100000

# Trending hashtags

# tags = ["fifa", "fifaworldcup", "fifaworldcup2022", "fifa2022", "fifaworldcupqatar2022", "qatarworldcup2022", "qatar", "qatar2022"]
# tags = ["worldcup","worldcup2022", "wcq", "worldcupqatar2022", "soccer", "football", "boycottqatar2022", "boycottfifaworldcup", "boycottwm2022", "boycotworldcup2022", "boycottqatar", "boycottfifa"]

# team hashtags
# tags = ["qtrecu", "qatecu", "qatarvsecuador", "ecuador", "qat", "eng", "engira", "engusa", "engwal", "usawal", "fraaus", "belcan", "breser"]

# match hashtags
tags = ["porarg", "usmnt", "onlyforward", "solopalante", "worldcupfantasy", "selecciónmayor", "vesteabandeira", "weartheflag", "threelions", "atlaslions", "dimamaghrib", "orlovi", "fss", "blackstars", "arbenybyd", "srcebije", "romsvn", "goats", "cr7", "messi"]

for t in tags:
    d = datetime(2022,1,1)
    while d < datetime.today():
        
        nextdate = d + relativedelta(months=+1)
        startdate = d.strftime("%Y-%m-%d")
        enddate = nextdate.strftime("%Y-%m-%d")
    
        p = "./data/teams/"+ t
        isExist = os.path.exists(p)
        
        if not isExist:
            os.makedirs(p)
            
        fname = p + "/tweet_" + tag + "_" + startdate + "_" + enddate + ".json"
        command = "snscrape --jsonl --max-results " + str(res_limit) + " --since " + startdate + " twitter-hashtag \"" + t + " until:" + enddate + "\"" + " > " + fname
        d = d + relativedelta(months=+1)
        
        print(command)
        os.system(command)
```
